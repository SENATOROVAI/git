"""Git и организация программных проектов."""

### Установка Git и проверка версии
'git --version' - чтобы узнать, что GIT установлен, а также версию.

### Работа с Git

1. Создание репозитория:
    - 'git init' - инициализирует новый репозиторий Git в текущей папке;
    - 'git clone' - для копирования существующего.
2. Добавление файлов для отслеживания:
    - 'git add <имя_файла>';
    - 'git add . ' - для добавления всех файлов.
3. Сохранение изменений:
    - 'git commit -am "<сообщение коммита>"'
4. Проверка статуса файлов:
    - 'git status' - показывает состояние файлов.
5. Просмотр изменений перед коммитом:
    - 'git diff' - показать все несохраненные изменения
6. Удаление и восстановление файлов:
    - 'git rm file.txt' - удалить файл;
    - 'git restore file.txt' - восстановить файл.
7. Переименование и перемещение файлов:
    - 'git mv старый.txt новый.txt' - переименование;
    - 'git mv файл.txt новая_папка/файл.txt' - перемещение;
    - 'git mv старый_путь/старый.txt новый_путь/новый.txt' - комбинированные операции.
8. Просмотр журнала коммитов (git log):
    - 'git log' - полный журнал;
    - 'git log --oneline' - сокращенный формат;
    - 'git log -n 3' - последние 3 коммита;
    - 'git show хеш:файл.txt' - содержимое файла в коммите.
9. Отмена несохраненных изменений:
    - 'git restore file.txt' - отменить изменения в одном файле;
    - 'git checkout . ' - отменить все изменения во всех файлах.
10. Деиндексирование файлов (unstage):
    - 'git restore --staged file.txt' - убрать файл из индекса.
11. Отмена последних коммитов (git revert):
    - 'git revert -n HEAD~3..HEAD' - отменить последние 3 коммита.
12. Возврат к определенной версии файла:
    - 'git show хеш:файл.py' - просмотр старой версии;
    - 'git checkout хеш -- файл.py' - восстановление файла.
13. Работа с конфиденциальными данными в Git:
    - Предотвращение утечек:
        # .gitignore
        secrets.txt
        confidential.py
        *.key
        *.password
    - 'git filter-branch' - если данные уже попали в репозиторий.
13. Отправка локального репозитория:
    - 'git remote add origin https://github.com/username/repo.git' - привязка удаленного репозитория;
    - 'git push -u origin master' - первая отправка;
    - 'git push' - последующие отправки.

### Справка по командам
    
    - 'git help' - получить справку по командам в GitHub;
    - 'git help <команда>' - получить справку по определённой команде;
    - 'man git-config' - просмотреть страницу руководства по конкретной команде Git.
